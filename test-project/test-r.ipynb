* Use [readxl](https://readxl.tidyverse.org/) to read the excel data into data frames.
* Use [plyr](https://rdrr.io/cran/plyr/) to combine data frames replacing unknown columns by NA.

```r
install.packages("readxl")
library(readxl)
library(plyr)
```

Read the excel data into a dataframe using the read_excel function from readxl

```r
animal_db_data <- data.frame(read_excel("animal_db.xlsx"))
animal_db_data
```

```r
animal_wiki_data <- data.frame(read_excel("animal_wiki.xlsx"))
animal_wiki_data
```

Combine the data by using the rbind.fill function from plyr

```r
total <- rbind.fill(animal_db_data, animal_wiki_data)
total
```

```r
plot(total)
```
